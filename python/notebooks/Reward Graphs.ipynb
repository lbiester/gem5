```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## Positive/Negative Reward fn

```python
use_window = 128

x = np.linspace(1,use_window,use_window)
y = np.ones(use_window)
y[-1] = -1
plt.figure(figsize=(8,2.5))
plt.title("Basic Reward Function")
sns.lineplot(x,y)
plt.show()
```

## Linear Reward Function

```python
use_window = 128

x = np.linspace(1,use_window,use_window)
y = [(use_window - x) / use_window for x in range(use_window)]
y[-1] = -1
plt.figure(figsize=(8,2.5))
plt.title("Linear Reward Function")
sns.lineplot(x,y)
plt.show()
```

## Curved Reward fn

```python
use_window = 128

x = np.linspace(1,use_window,use_window)
y = [1/(np.sqrt(2*np.pi*8**2)) * np.exp(-(i-36)**2/(2*8**2)) for i in x]
# shift and scale (rewards <20 and > 50 are negative)
y = [(i-0.01)*100 for i in y]
# adjust rewards beyond 60 so that they're increasingly negative
for i in range(36+8*3,use_window):
    y[i] = y[i-1]*1.02
plt.figure(figsize=(8,2.5))
plt.title("Curved Reward Function")
sns.lineplot(x,y)
plt.show()
```
